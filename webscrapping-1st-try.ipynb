```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib
from bs4 import BeautifulSoup
import datetime
import ast
import csv
from IPython.display import HTML

%matplotlib inline
```

```
HTML('https://ca.indeed.com/jobs?q=data+scientist&l=Toronto')
```

```
url = urllib.request.urlopen('https://ca.indeed.com/jobs?q=data+scientist&l=Toronto')
content = url.read()
#soup = BeautifulSoup(content, 'html5lib')
soup = BeautifulSoup(content, 'lxml')
print(soup)
```

```
links = []

for h2Tag in soup.find_all('h2',{'class': 'title'}):
    for aTag in h2Tag.find_all('a'):
        links.append(aTag.get('href'))
            
links = links[:-1]
links
```

```
raw_url = 'https://ca.indeed.com' + links[0]
raw_url
```

```
url = urllib.request.urlopen('https://ca.indeed.com' + links[0])

content = url.read()
soup = BeautifulSoup(content, 'html5lib')

description = soup.find('div',{'id': 'jobDescriptionText'}).text
description
```
